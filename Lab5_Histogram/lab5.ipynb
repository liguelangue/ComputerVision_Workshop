```python
import cv2
import numpy as np

# Load the image in grayscale mode
img = cv2.imread('flower.png', 0)
```

```python
# Calculate the Histogram
hist = cv2.calcHist([img], [0], None, [256], [0, 256])

import matplotlib.pyplot as plt
# Plot the grayscale histogram
plt.plot(hist)
plt.title('Grayscale Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
```

```python
img_color = cv2.imread("flower.png")

# Initialize colors for BGR channels
colors = ('b', 'g', 'r')

# Loop through each color channel (Blue, Green, Red)
for i, col in enumerate(colors):
    # Calculate the histogram for each channel
    hist_color = cv2.calcHist([img_color], [i], None, [256], [0, 256])
    # Plot the histogram for the current channel
    plt.plot(hist_color, color=col)

plt.title('Color Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
```
