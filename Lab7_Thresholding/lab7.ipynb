```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image in grayscale
image = cv2.imread('photo.png', cv2.IMREAD_GRAYSCALE)# Show the image

plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.show()
```

```python
# Apply simple thresholding
_, thresh_simple = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

# Show the image
plt.imshow(thresh_simple, cmap='gray')
plt.title('Simple Thresholding')
plt.show()
```

```python
# Apply adaptive mean thresholding
thresh_adaptive_mean = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
# Apply adaptive Gaussian thresholding
thresh_adaptive_gaussian = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
# Show both results
plt.subplot(1, 2, 1), plt.imshow(thresh_adaptive_mean, cmap='gray')
plt.title('Adaptive Mean Thresholding')
plt.subplot(1, 2, 2), plt.imshow(thresh_adaptive_gaussian, cmap='gray')
plt.title('Adaptive Gaussian Thresholding')
plt.show()
```

```python
# Change the block size of adaptive mean thresholding
# Apply adaptive mean thresholding
thresh_adaptive_mean_11 = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
# Apply adaptive mean thresholding
thresh_adaptive_mean_7 = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 7, 2)
# Show both results
plt.subplot(1, 2, 1), plt.imshow(thresh_adaptive_mean_11, cmap='gray')
plt.title('A_Mean_T with block size 11')
plt.subplot(1, 2, 2), plt.imshow(thresh_adaptive_mean_7, cmap='gray')
plt.title('A_Mean_T with block size 7')
plt.show()
```

```python
# Apply adaptive Gaussian thresholding
thresh_adaptive_gaussian_11 = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
# Apply adaptive Gaussian thresholding
thresh_adaptive_gaussian_7 = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 7, 2)
# Show both results
plt.subplot(1, 2, 1), plt.imshow(thresh_adaptive_gaussian_11, cmap='gray')
plt.title('A_Gaussian_T with block size 11')
plt.subplot(1, 2, 2), plt.imshow(thresh_adaptive_gaussian_7, cmap='gray')
plt.title('A_Gaussian_T with block size 7')
plt.show()
```

```python
# Apply Otsus binarization
_, thresh_otsu = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
# Show the result
plt.imshow(thresh_otsu, cmap='gray')
plt.title('Otsus Binarization')
plt.show()
```

```python
# Load image in grayscale
image = cv2.imread('einstein.png', cv2.IMREAD_GRAYSCALE)# Show the image
# Apply simple thresholding
_, thresh_simple = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
# Apply adaptive mean thresholding
thresh_adaptive_mean = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
# Apply adaptive Gaussian thresholding
thresh_adaptive_gaussian = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
# Apply Otsus binarization
_, thresh_otsu = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
# Show both results
plt.subplot(3, 2, 1), plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.subplot(3, 2, 2), plt.imshow(thresh_simple, cmap='gray')
plt.title('Simple  Thresholding')
plt.subplot(3, 2, 3), plt.imshow(thresh_adaptive_mean, cmap='gray')
plt.title('Adaptive Mean Thresholding')
plt.subplot(3, 2, 4), plt.imshow(thresh_adaptive_gaussian, cmap='gray')
plt.title('Adaptive Gaussian Thresholding')
plt.subplot(3, 2, 5), plt.imshow(thresh_otsu, cmap='gray')
plt.title("Otsu's binarization Thresholding")
plt.show()
```
